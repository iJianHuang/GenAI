# Spam

#### Load CSV

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report

df = pd.read_csv('/content/spam.csv')
df.head()
```

```
df.Category.value_counts()
```

```
df.Category.value_counts() / len(df) * 100
```

#### Add column 'spam'

```
df['spam'] = df['Category'].apply(lambda x: 1 if x=='spam' else 0)
df.head()
```

```
df.shape
```

#### Train test split

```
# from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df.Message, df.spam, test_size=0.25)
```

```
x_train.shape
```

```
x_test
```

```
x_train[:4]
```

```
y_train[:4]
```

#### Create bag of words representation using **CountVectorization**

```
v = CountVectorizer()
x_train_cv = v.fit_transform(x_train.values)
x_train_cv.toarray()[:2]
```

```
x_train_cv.toarray().shape
```

```
display(v.get_feature_names_out())
```

```
v.get_feature_names_out()[1000]
```

```
v.vocabulary_
```

```
x_train_np = x_train_cv.toarray()
x_train_np
```

```
np.where(x_train_np[0]!=0)
```

```
# Swap keys and values using dictionary comprehension
swapped_dict = {value: key for key, value in v.vocabulary_.items()}

print(swapped_dict)
```

```
for i in np.where(x_train_np[0]!=0)[0]:
  print(swapped_dict[i])
```

```
x_train.iloc[0]
```

```
v.get_feature_names_out()[885]
```

#### **Naive Bayes Classifier**

```
# from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(x_train_cv, y_train)
```

```
x_test_cv = v.transform(x_test.values)
y_pred = model.predict(x_test_cv)
```

```
print (classification_report(y_test, y_pred))
```

#### **Test on random data points**

```
message = {"Spook-tacular savings alert – $800 OFF + up to $100 Alaska bonus"}
message_cv = v.transform(message)
model.predict(message_cv)
```
